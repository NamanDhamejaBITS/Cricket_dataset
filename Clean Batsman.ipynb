```python
def clean_batsman(file):
    fe=f"data/{file}.csv"
    df=pd.read_csv(fe)
    df.drop(labels="Unnamed: 0",axis=1,inplace=True) #Remove the Unnamed column
    a=[]
    # Remove entries with DNB,TDNB and absent
    for i in range(0,len(df)):
        if( df["4s"][i] =='-'):
            df["4s"][i] = 0
        if( df["6s"][i] =='-'):
            df["6s"][i] = 0
        if(df.Runs[i]=="DNB" or df.Runs[i]=="TDNB" or df.Runs[i]=="absent"):
            a.append(i)
    df.drop(a,inplace=True)
    df.reset_index(drop=True, inplace=True)
    df["Runs"] = df["Runs"].str.replace(r'*', '').astype(int)
    # Replace '*' with space, to convert the column to numbers
    df["SR"] = df["SR"].str.replace('-', '0').astype(float)
    df["BF"] = df["BF"].str.replace('-', '0').astype(float)
    return(df)
```

```python
data=clean_batsman("kohli")
```

```python
def ground_details(l,df):
        ru=0
        s4=0
        s6=0
        cnt=0
        sr=0
        bf=0
        for i in range(0, len(df)):
            if(df["Ground"][i]==l):
                ru=ru+int(df["Runs"][i])
                s4=s4+int(df["4s"][i])
                s6=s6+int(df["6s"][i])
                cnt=cnt+1
                sr=sr+int(df["SR"][i])
                bf=bf+int(df["BF"][i])
        return(ru,s4,s6,cnt,round(sr/cnt,2),bf)
```

```python
def prep_dataset_ground(batsman):
    ndf=pd.DataFrame(columns=["Batsman","Ground","Runs","4s","6s","No. of Matches Played","Average Strike Rate","Balls Faced"])
    
    for bat in batsman:
        #print(str(x))
        df=clean_batsman(str(bat))
        li=list(set(df["Ground"]))
        for gnd in li:
            ru,s_4,s_6,nmf,avg_sr,bf=ground_details(gnd,df)
            add={"Batsman":bat.capitalize(),"Ground":gnd,"Runs":ru,"4s":s_4,"6s":s_6,"No. of Matches Played":nmf,"Average Strike Rate":avg_sr,"Balls Faced":bf}
            ndf=ndf.append(add,ignore_index=True)
    return(ndf)        
            
```

```python
batsman_ground=prep_dataset_ground(batsman)
```
