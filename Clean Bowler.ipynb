```python
bowlers={'ashwin',
 'ashwin1',
 'badree',
 'kumble',
'mitchell',
 'kumble1',
 'kumblesp',
 'malinga',
 'malinga1','mendis','murali',
 'narine','southee',
 'steyn','warne'}
```

```python
def clean_bowler(file):
    fe=f"data/{file}.csv"
    df=pd.read_csv(fe)
    df.drop(labels="Unnamed: 0",axis=1,inplace=True) #Remove the Unnamed column
    df["Bowls_Bowled"]=df["Overs"]*6
    #df["Bowls_Bowled"]=df["Bowls_Bowled"].astype(int)
    a=[]
    # Remove entries with DNB,TDNB and absent
    for i in range(0,len(df)):
        
        if(df.Overs[i]=="DNB" or df.Overs[i]=="TDNB" or df.Overs[i]=="absent" or df.Overs[i]=="-"):
            a.append(i)
    df.drop(a,inplace=True)
    df.reset_index(drop=True, inplace=True)
    df["Overs"]=df["Overs"].astype(float)
    for i in range(0,len(df)):
        num = float(df["Overs"][i])
        dec,whl = math.modf(num)
        #print(int(round(dec,1)*10+ whl*6))
        df["Bowls_Bowled"][i]=int(round(dec,1)*10+ whl*6)
        if( df["Wkts"][i] =='-'):
            df["Wkts"][i] = 0
        if( df["Econ"][i] =='-'):
            df["Econ"][i] = df["Runs"][i]/df["Overs"][i]
    df.reset_index(drop=True, inplace=True)
#     df["Runs"] = df["Runs"].str.replace(r'*', '').astype(int)
#     # Replace '*' with space, to convert the column to numbers
#     df["SR"] = df["SR"].str.replace('-', '0').astype(float)
#     df["BF"] = df["BF"].str.replace('-', '0').astype(float)
    return(df)
```

```python
clean_bowler("steyn")
```

```python
def over_clean(num):
    dec,whl = math.modf(num)
    if(round(dec,1) >= 0.6):
        num = round(num,1)+(1-round(dec,1))
    return num


def bowl_calculator(over):
    dec,whl = math.modf(over)
    bf=int(round(dec,1)*10+ whl*6)
    return bf
```

```python
def ground_details(l,df):
        ru=0
        over=0
        wkts=0
        cnt=0
        ec=0
        bf=0
        for i in range(0, len(df)):
            if(df["Ground"][i]==l):
                ru=ru+int(df["Runs"][i])
                over=over+float(df["Overs"][i])
                wkts=wkts+int(df["Wkts"][i])
                cnt=cnt+1
                ec=ec+float(df["Econ"][i])
                over=over_clean(over)
                bf=bowl_calculator(over)
        return(ru,round(over,1),bf,int(cnt),round(ec/cnt,2),wkts)
```

```python
def prep_dataset_ground(bowler):
    ndf=pd.DataFrame(columns=["Bowler","Ground","Overs","Balls Bowled","Wickets","No. of Matches Played","Average Economy","Runs"])
    
    for bowl in bowler:
        #print(str(x))
        df=clean_bowler(str(bowl))
        li=list(set(df["Ground"]))
        for gnd in li:
            ru,over,bf,nmf,avg_ec,wkts=ground_details(gnd,df)
            add={"Bowler":bowl.capitalize(),"Ground":gnd,"Runs":ru,"Overs":over,"Balls Bowled":bf,"No. of Matches Played":nmf,"Average Economy":avg_ec,"Wickets":wkts}
            ndf=ndf.append(add,ignore_index=True)
    return(ndf)        
            
```

```python
prep_dataset_ground(bowlers)
```

```python
bowler_ground=prep_dataset_ground(bowlers)
```

```python
bowler_ground.to_csv('Bowler_ground.csv')
```

```python
def opposition_details(l,df):
        ru=0
        over=0
        wkts=0
        cnt=0
        ec=0
        bf=0
        for i in range(0, len(df)):
            if(df["Opposition"][i]==l):
                ru=ru+int(df["Runs"][i])
                over=over+float(df["Overs"][i])
                wkts=wkts+int(df["Wkts"][i])
                cnt=cnt+1
                ec=ec+float(df["Econ"][i])
                over=over_clean(over)
                bf=bowl_calculator(over)
        return(ru,round(over,1),bf,int(cnt),round(ec/cnt,2),wkts)
```

```python
def prep_dataset_opposition(bowler):
    ndf=pd.DataFrame(columns=["Bowler","Opposition","Overs","Balls Bowled","Wickets","No. of Matches Played","Average Economy","Runs"])
    
    for bowl in bowler:
        #print(str(x))
        df=clean_bowler(str(bowl))
        li=list(set(df["Opposition"]))
        for gnd in li:
            ru,over,bf,nmf,avg_ec,wkts=opposition_details(gnd,df)
            add={"Bowler":bowl.capitalize(),"Opposition":gnd,"Runs":ru,"Overs":over,"Balls Bowled":bf,"No. of Matches Played":nmf,"Average Economy":avg_ec,"Wickets":wkts}
            ndf=ndf.append(add,ignore_index=True)
    return(ndf)        
            
```

```python
prep_dataset_opposition(bowlers)
```

```python
bowler_opposition=prep_dataset_opposition(bowlers)
```

```python
bowler_opposition.to_csv('Bowler_Opposition.csv')
```
