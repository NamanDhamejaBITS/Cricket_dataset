# Clean BATSMAN Function

```python
def clean_batsman(file):
    fe=f"data/{file}.csv"
    df=pd.read_csv(fe)
    df.drop(labels="Unnamed: 0",axis=1,inplace=True) #Remove the Unnamed column
    a=[]
    # Remove entries with DNB,TDNB and absent
    for i in range(0,len(df)):
        if( df["4s"][i] =='-'):
            df["4s"][i] = 0
        if( df["6s"][i] =='-'):
            df["6s"][i] = 0
        if(df.Runs[i]=="DNB" or df.Runs[i]=="TDNB" or df.Runs[i]=="absent"):
            a.append(i)
    df.drop(a,inplace=True)
    df.reset_index(drop=True, inplace=True)
    df["Runs"] = df["Runs"].str.replace(r'*', '')
    # Replace '*' with space, to convert the column to numbers
    return(df)
```

```python
def ground_details(l,df):
        ru=0
        s4=0
        s6=0
        cnt=0
        sr=0
        bf=0
        for i in range(0, len(df)):
            if(df["Ground"][i]==l):
                ru=ru+int(df["Runs"][i])
                s4=s4+int(df["4s"][i])
                s6=s6+int(df["6s"][i])
                cnt=cnt+1
                sr=sr+float(df["SR"][i])
                bf=bf+int(df["BF"][i])
        return(ru,s4,s6,cnt,round(sr/cnt,2),bf)
```
